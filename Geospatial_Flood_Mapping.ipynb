3A. Storing classification in CSV file

```python
import numpy as np
from netCDF4 import Dataset
import random
import os

def check_extent(file_name, lat_val, lon_val):
    read_data = Dataset(file_name, 'r')
    var = list(read_data.variables.keys())

    read_lat = read_data.variables['lat'][:]
    read_lon = read_data.variables['lon'][:]
    im = read_data.variables['Band1'][:]

    lon_arr = read_lon[:]
    lat_arr = read_lat[:]

    diff_arr_x = np.absolute(lon_arr - lon_val)
    id_x = diff_arr_x.argmin()

    diff_arr_y = np.absolute(lat_arr - lat_val)
    id_y = diff_arr_y.argmin()

    x = im[id_y][id_x]

    try:
        x = int(x)
    except:
        x = 0

    return x

def read_netCDF_data(file_name, extent_file, output_file):
    num_cells = []
    num_each_class = 87

    read_data = Dataset(file_name, 'r')
    var = list(read_data.variables.keys())

    read_lat = read_data.variables['lat'][:]
    read_lon = read_data.variables['lon'][:]
    im = read_data.variables['Band1'][:]

    lon_arr = read_lon[:]
    lat_arr = read_lat[:]

    lon_positive = []
    num_positive = 0
    lon_negative = []
    num_negative = 0

    for i in range(len(lat_arr)):
        lon_positive.append([])
        lon_negative.append([])
        for j in range(len(lon_arr)):
            temp_var = im[i][j]
            x = check_extent(extent_file, lat_arr[i], lon_arr[j])
            if not np.isnan(temp_var) and int(temp_var) == 1 and x == 1:
                lon_positive[i].append(lon_arr[j])
                num_positive += 1
            elif not np.isnan(temp_var) and int(temp_var) == 0 and x == 1:
                lon_negative[i].append(lon_arr[j])
                num_negative += 1

    num_each_lat_positive = []
    num_each_lat_negative = []

    coeff_pos = 1
    while sum(num_each_lat_positive) < num_each_class - 5 or sum(num_each_lat_positive) > num_each_class:
        num_each_lat_positive = []
        for i in range(len(lon_positive)):
            num_each_lat_positive.append(int(len(lon_positive[i]) * num_each_class * coeff_pos / num_positive))
        if sum(num_each_lat_positive) == 0:
            coeff_pos += 1
        else:
            coeff_pos += (num_each_class - sum(num_each_lat_positive)) / sum(num_each_lat_positive)

    coeff_neg = 1
    while sum(num_each_lat_negative) < num_each_class - 5 or sum(num_each_lat_negative) > num_each_class:
        num_each_lat_negative = []
        for i in range(len(lon_negative)):
            num_each_lat_negative.append(int(len(lon_negative[i]) * num_each_class * coeff_neg / num_negative))
        if sum(num_each_lat_negative) == 0:
            coeff_neg += 1
        else:
            coeff_neg += (num_each_class - sum(num_each_lat_negative)) / sum(num_each_lat_negative)

    num_each_lat_positive[num_each_lat_positive.index(max(num_each_lat_positive))] += num_each_class - sum(num_each_lat_positive)
    num_each_lat_negative[num_each_lat_negative.index(max(num_each_lat_negative))] += num_each_class - sum(num_each_lat_negative)

    data_pos = []
    data_neg = []

    for i in range(len(lat_arr)):
        arr_idx = list(np.arange(0, len(lon_positive[i])))
        selection_arr = random.sample(arr_idx, num_each_lat_positive[i])
        for j in range(len(selection_arr)):
            data_pos.append([lat_arr[i], lon_positive[i][int(selection_arr[j])]])

    for i in range(len(lat_arr)):
        arr_idx = list(np.arange(0, len(lon_negative[i])))
        selection_arr = random.sample(arr_idx, num_each_lat_negative[i])
        for j in range(len(selection_arr)):
            data_neg.append([lat_arr[i], lon_negative[i][int(selection_arr[j])]])

    write_csv(data_pos, data_neg, output_file)

    return 0

def write_csv(data_pos, data_neg, output_file):
    with open(output_file, 'w') as csv_file:
        str_write = ''
        for i in range(len(data_pos)):
            str_write += str(data_pos[i][1]) + ',' + str(data_pos[i][0]) + ',' + str(1) + '\n'
        for i in range(len(data_neg)):
            str_write += str(data_neg[i][1]) + ',' + str(data_neg[i][0]) + ',' + str(0) + '\n'
        csv_file.write(str_write)
    return 0

file_path = '/content/'
input_file = os.path.join(file_path, 'Brahmaputra_Total_Flooded.nc')
extent_file = os.path.join(file_path, 'Brahmaputra_Area.nc')
output_file = os.path.join(file_path, 'data.csv')

read_netCDF_data(input_file, extent_file, output_file)
```

Creating the final dataset

```python
import numpy as np
import pandas as pd
from netCDF4 import Dataset
import random

def read_netCDF_data(file_name, lon, lat, out_var='Band1'):
    z = 0
    read_data = Dataset(file_name, 'r')
    read_lat = read_data.variables['lat'][:]
    read_lon = read_data.variables['lon'][:]
    im = read_data.variables[out_var][:]

    lon_arr = read_lon[:]
    lat_arr = read_lat[:]

    diff_arr_x = np.absolute(lon_arr - lon)
    id_x = diff_arr_x.argmin()

    diff_arr_y = np.absolute(lat_arr - lat)
    id_y = diff_arr_y.argmin()

    z = im[id_y][id_x]

    return z

def write_csv(file_name, data_arr):
    with open(file_name, 'w') as csv_file:
        for data in data_arr:
            str_write = ','.join(map(str, data)) + '\n'
            csv_file.write(str_write)

def read_data_create_array(Basin_name, input_layers):
    path_coords_read = f'/content/data.csv'
    path_input_layers = '/content'

    matrix_coords = pd.read_csv(path_coords_read, header=None).to_numpy()
    matrix_coords = matrix_coords.astype('float64')
    matrix_coords = matrix_coords[matrix_coords[:, 2].argsort()]

    mode_val = [1]
    mode_itr = 0
    prev_val = matrix_coords[0, 2]

    for i in range(1, len(matrix_coords)):
        if matrix_coords[i, 2] == prev_val:
            mode_val[mode_itr] += 1
        else:
            mode_itr += 1
            prev_val = matrix_coords[i, 2]
            mode_val.append(1)

    validation_data_points = np.round(np.multiply(mode_val, 0.2), decimals=0).astype('int')
    training_data_points = np.subtract(mode_val, validation_data_points)

    randomize_itr_start = 0

    for i in range(len(mode_val)):
        randomize_itr_stop = int(int(mode_val[i]) + randomize_itr_start)
        np.random.shuffle(matrix_coords[randomize_itr_start:randomize_itr_stop])
        randomize_itr_start += int(mode_val[i])

    itr = 0
    training_data = []

    for i in range(len(training_data_points)):
        for j in range(training_data_points[i]):
            data = matrix_coords[itr + j, :3].tolist()  # Latitude, Longitude, Class Label
            for k in range(len(input_layers)):
                file_name = f'{path_input_layers}/{input_layers[k]}.nc'  # Read NetCDF file
                z = read_netCDF_data(file_name, matrix_coords[itr + j, 0], matrix_coords[itr + j, 1], 'Band1')
                data.append(z)

            training_data.append(data)

        itr += mode_val[i]

    np.random.shuffle(training_data)

    itr = training_data_points[0]
    validation_data = []

    for i in range(len(validation_data_points)):
        for j in range(validation_data_points[i]):
            data = matrix_coords[itr + j, :3].tolist()  # Latitude, Longitude, Class Label
            for k in range(len(input_layers)):
                file_name = f'{path_input_layers}/{input_layers[k]}.nc'  # Read NetCDF file
                z = read_netCDF_data(file_name, matrix_coords[itr + j, 0], matrix_coords[itr + j, 1], 'Band1')
                data.append(z)

            validation_data.append(data)

        itr += mode_val[i]

    np.random.shuffle(validation_data)

    # Write datasets to CSV
    path_save = '/content/'
    training_file_name = f'{path_save}{Basin_name}_training.csv'
    validation_file_name = f'{path_save}{Basin_name}_validation.csv'
    write_csv(training_file_name, training_data)
    write_csv(validation_file_name, validation_data)

    print(f'Training dataset created: {training_file_name}')
    print(f'Validation dataset created: {validation_file_name}')

basin = 'Brahmaputra'
input_layers = ['Accumulated_Precipitation', 'Aspect', 'Dist_Stream', 'Filled_DEM', 'Precipitation',
                'SPI', 'STI', 'Slope', 'Streamflow', 'TRI', 'TWI', 'Total_Curvature']

read_data_create_array(basin, input_layers)
```

Preprocessing the data

```python
import pandas as pd

training_data = pd.read_csv('/content/Brahmaputra_training.csv', header=None)
testing_data = pd.read_csv('/content/Brahmaputra_validation.csv', header=None)

print("Training Data:")
print(training_data.head())

print("\nTesting Data:")
print(testing_data.head())

print("\nMissing Values in Training Data:")
print(training_data.isnull().sum())

print("\nMissing Values in Testing Data:")
print(testing_data.isnull().sum())

print("\nStatistical Summary of Training Data:")
print(training_data.describe())

print("\nStatistical Summary of Testing Data:")
print(testing_data.describe())

print("Number of occurrences of each target variable in the training dataset:")
print(training_data[2].value_counts())
```

```python
import pandas as pd
training_df = pd.DataFrame(training_data)
correlation_matrix = training_df.corr()
correlation_with_target = correlation_matrix.iloc[2, :-1]
print("Correlation of each attribute with the target variable in descending order:")
print(correlation_with_target.abs().sort_values(ascending=False))

```

```python
from sklearn.preprocessing import RobustScaler
import pandas as pd
X = training_df.iloc[:, 3:]
y = training_df.iloc[:, 2]

scaler = RobustScaler()
X_normalized = scaler.fit_transform(X)
X_normalized_df = pd.DataFrame(X_normalized, columns=X.columns)
normalized_data = pd.concat([training_df.iloc[:, :3], X_normalized_df], axis=1)
correlation_matrix_normalized = normalized_data.corr()
correlation_with_target_normalized = correlation_matrix_normalized.iloc[2, :-1]

print("Correlation of each normalized attribute with the target variable in descending order:")
print(correlation_with_target_normalized.abs().sort_values(ascending=False))
print("First 5 instances in the training dataset after normalization:")
print(normalized_data.head())

testing_df = pd.DataFrame(testing_data)
X_test = testing_df.iloc[:, 3:]
X_test_normalized = scaler.transform(X_test)
X_test_normalized_df = pd.DataFrame(X_test_normalized, columns=X_test.columns)
normalized_testing_data = pd.concat([testing_df.iloc[:, :3], X_test_normalized_df], axis=1)
print("\nFirst 5 instances in the testing dataset after normalization:")
print(normalized_testing_data.head())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_curve, auc
import matplotlib.pyplot as plt

X_train = normalized_data.iloc[:, 3:]
y_train = normalized_data.iloc[:, 2]

logreg = LogisticRegression()

logreg.fit(X_train, y_train)

y_test_prob = logreg.predict_proba(normalized_testing_data.iloc[:, 3:])[:, 1]

fpr, tpr, thresholds = roc_curve(normalized_testing_data.iloc[:, 2], y_test_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 8))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

y_test_pred = logreg.predict(normalized_testing_data.iloc[:, 3:])

accuracy = accuracy_score(normalized_testing_data.iloc[:, 2], y_test_pred)
precision = precision_score(normalized_testing_data.iloc[:, 2], y_test_pred)
recall = recall_score(normalized_testing_data.iloc[:, 2], y_test_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```

```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from netCDF4 import Dataset

training_data = pd.read_csv('/content/Brahmaputra_training1.csv')

X_train = training_data.iloc[:, 3:].values
y_train = training_data.iloc[:, 2].values

mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=5)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
mlp.fit(X_train_scaled, y_train)

flood_data = Dataset('/content/Brahmaputra_Total_Flooded.nc', 'r')

latitudes = flood_data.variables['lat'][:]
longitudes = flood_data.variables['lon'][:]
crs_data = flood_data.variables['crs'][:]

validation_data = pd.read_csv('/content/Brahmaputra_validation1.csv', header=None)

validation_latitudes = validation_data.iloc[:, 0].values
validation_longitudes = validation_data.iloc[:, 1].values

X_validation = validation_data.iloc[:, 3:].values
X_validation_scaled = scaler.transform(X_validation)
y_validation = mlp.predict(X_validation_scaled)

mlp_results = pd.DataFrame({
    'Latitude': validation_latitudes,
    'Longitude': validation_longitudes,
    'MLP_Prediction': y_validation
})

final_results = pd.DataFrame({
    'Latitude': mlp_results['Latitude'],
    'Longitude': mlp_results['Longitude'],
    'MLP_Prediction': mlp_results['MLP_Prediction'],
    'NearestLatitude': np.nan,
    'NearestLongitude': np.nan,
    'CRS': np.nan
})

def find_nearest_lat_lon(lat, lon, latitudes, longitudes):
    lat_idx = np.abs(latitudes - lat).argmin()
    lon_idx = np.abs(longitudes - lon).argmin()
    return latitudes[lat_idx], longitudes[lon_idx]

for index, row in final_results.iterrows():
    lat, lon = row['Latitude'], row['Longitude']
    nearest_lat, nearest_lon = find_nearest_lat_lon(lat, lon, latitudes, longitudes)
    final_results.at[index, 'NearestLatitude'] = nearest_lat
    final_results.at[index, 'NearestLongitude'] = nearest_lon
    final_results.at[index, 'CRS'] = crs_data

mlp_results.to_csv('/content/mlp_results.csv', index=False)

final_results.to_csv('/content/final.csv', index=False)

flood_data.close()
```
